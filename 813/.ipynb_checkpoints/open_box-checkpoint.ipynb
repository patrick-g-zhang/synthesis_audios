```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
import os
import re
import random
import time
import traceback
from IPython.display import Image
import pickle
from sklearn.manifold import TSNE
import tensorflow as tf
```

```python
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
```

```python
# load vae mean vectors for each utterance
style_embeddings = np.load("../../gst-tacotron/embedding_256.npy")
spk_index = np.load("../../gst-tacotron/spk_index.npy")
```

```python
X_embedded = TSNE(n_components=2).fit_transform(style_embeddings)
```

```python
x_min, x_max = X_embedded.min(0), X_embedded.max(0)
X_norm = (X_embedded - x_min) / (x_max - x_min)  # 归一化
```

```python
for i in range(X_norm.shape[0]):
    plt.text(X_norm[i, 0], X_norm[i, 1], str(spk_index[i]),
                 color=plt.cm.Set1(spk_index[i] / 10.),
                 fontdict={'weight': 'bold', 'size': 4})
```

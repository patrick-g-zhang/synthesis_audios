```python
from sklearn.manifold import TSNE
import numpy as np
import os
import re
import random
import time
import traceback
from IPython.display import Image
```

```python
import pickle
```

```python
text_emb=tf.contrib.framework.load_variable('/home/gyzhang/projects/gst-tacotron/logs-tacotron/cusent/model.ckpt-154000','model/inference/text_embedding')
```

```python
_pad        = '_'
_eos        = '~'
pkl_file = open('/home/gyzhang/projects/gst-tacotron/text/final_set', 'rb')
final_set = pickle.load(pkl_file)
pkl_file.close()
pkl_file = open('/home/gyzhang/projects/gst-tacotron/text/initial_set', 'rb')
initial_set = pickle.load(pkl_file)
pkl_file.close()

symbols = list(initial_set)+list(final_set)
symbols = [re.split('\_',symbol)[1] for symbol in symbols]
```

```python
symbols=[_pad,_eos] +symbols
```

```python
X = np.array(text_emb)
X_embedded = TSNE(n_components=2).fit_transform(X)
```

```python
index_list=[]
for num,symbol in enumerate(symbols):
    if symbol[-1] == '1' or not symbol[-1].isdigit():
        index_list.append(num)
        print(num, symbol)
```

```python
plt.rcParams['savefig.dpi'] = 300 #图片像素
plt.rcParams['figure.dpi'] = 300 #分辨率
```

```python
X_sub_embed = X_embedded[index_list,:]
```

```python
x_min, x_max = X_embedded.min(0), X_embedded.max(0)
X_norm = (X_sub_embed - x_min) / (x_max - x_min)  # 归一化
for i in range(X_sub_embed.shape[0]):
    if i<=20:
        plt.text(X_norm[i, 0], X_norm[i, 1],s=symbols[index_list[i]],bbox=dict(facecolor='red',alpha=0.5),fontsize=4)
    else:
        plt.text(X_norm[i, 0], X_norm[i, 1],s=symbols[index_list[i]],fontsize=4)
plt.xlim(0.0,0.6)
plt.ylim(0.0,0.8)
```
